```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([i for i in range(1, 273+1)])
y = np.array([-i for i in range(1, 199)])
len(x)
```

```python
len(y)
```

```python
x = x[-198:]
```

```python
rz = np.tan(x)/np.tan(y) * np.tan(y)/np.cos(x)

plt.plot(x, y)
plt.plot(x, rz)
plt.plot(rz, x)
plt.plot(y, rz)
plt.plot(rz, y)










```
